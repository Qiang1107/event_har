```python
import matplotlib.pyplot as plt
import numpy as np

# 原始传感器时间点和对应值
t_original = np.array([0, 1, 3, 4])  # 时间
y_original = np.array([0, 2, 1, 3])  # 传感器值

# 新的统一时间轴（更多采样点）
t_uniform = np.linspace(0, 4, 20)

# 线性插值
y_interpolated = np.interp(t_uniform, t_original, y_original)

# 绘图
plt.figure(figsize=(8, 4))
plt.plot(t_original, y_original, 'o-', label='原始数据', markersize=8)
plt.plot(t_uniform, y_interpolated, '--', label='线性插值后的数据')
plt.xlabel('时间')
plt.ylabel('传感器值')
plt.title('线性插值示意图')
plt.legend()
plt.grid(True)
plt.show()
```
